## Student Course Enrollment System
### Problem Statement: ###

###### You are required to build a simple Student Course Enrollment System using Python. The system will allow users to: ######

###### 1.Register students with their names and a list of courses they want to enroll in. ######

###### 2.Display all registered students with their enrolled courses. ######

###### 3.Count and display how many students have enrolled in each course. ######

```python
available_courses ={course.title() for course in {'Python', 'DSA', 'Advance DSA', 'Full Stack', 'Data Visualization'}}
student_going_to_register=int(input("How Many Students are going to register ?  "))
student_data=[]
lsit_of_enrolled_course=[]

for i in range(student_going_to_register):
    print(f"\n Registering Students {i+1}")
    name=input("Enter your name: ").strip().capitalize()

    select_courses = input("Enter Course(s) You Want to Enroll (comma separated): ").strip()
    select_course_list = [course.strip().title() for course in select_courses.split(',')]

    # Validate each course
    while any(course not in available_courses for course in select_course_list):
        print("Invalid course(s)! Please choose from the available list.")
        select_courses = input("Enter Course(s) You Want to Enroll (comma separated): ").strip()
        select_course_list = [course.strip().title() for course in select_courses.split(',')]

    # For each course student enrolled, create a record and append
    for course in select_course_list:
        student_record = (name, course)
        student_data.append(student_record)
        lsit_of_enrolled_course.append(course)


print("\n Registered Students are !! ")
for student in student_data:
    print(f"Name of the Student is: {student[0]}|You Enrolled in: {student[1]}")
    
def count_enrollments(student_data):
    course_counts = {}
    for student in student_data:
        course = student[1]
        course_counts[course] = course_counts.get(course, 0) + 1

    if not course_counts:
        print("\nNo enrollments yet.")
        return

    print("\nEnrollments per course:")
    for course, count in sorted(course_counts.items()):
        print(f"{course}: {count} student{'s' if count != 1 else ''}")

count_enrollments(student_data)


```

Contact Book Management System
Problem Statement:

1.Create a Contact Book Management System using Python. The system should allow the user to:

2.Add a new contact with name, phone number, and email.

3.Search for a contact by name.

4.Update the phone number or email of an existing contact.

5.Delete a contact.

6.Display all saved contacts.




```python
# contact_book_simple

contacts = {}  

def add_contact():
    name = input("Enter name: ").strip()
    if not name:
        print("Name can’t be blank.")
        return

    if name in contacts:
        print("Contact already exists.")
        return

    phone = input("Enter phone: ").strip()
    email = input("Enter email: ").strip()
    contacts[name] = [phone, email]
    print(f"  Added '{name}'.")

def search_contact():
    name = input("Search contact name: ").strip()
    if name in contacts:
        phone, email = contacts[name]
        print(f"  {name}: {phone} | {email}")
    else:
        print("No such contact.")

def update_contact():
    name = input("Name to update: ").strip()
    if name not in contacts:
        print("Not found.")
        return

    old_phone, old_email = contacts[name]
    new_phone = input(f"New phone (Enter to keep {old_phone}): ").strip()
    new_email = input(f"New email (Enter to keep {old_email}): ").strip()

    if new_phone:
        contacts[name][0] = new_phone
    if new_email:
        contacts[name][1] = new_email

    print(f"Updated '{name}'.")

def delete_contact():
    name = input("Name to delete: ").strip()
    if name in contacts:
        del contacts[name]
        print(f"Deleted '{name}'.")
    else:
        print("Not found.")

def display_all():
    if not contacts:
        print("No contacts saved.")
        return

    print("\nAll contacts:")
    for name, (phone, email) in contacts.items():
        print(f"{name}: {phone} | {email}")
    print()

def main_menu():
    while True:
        print("\n----Contact Book----")
        print("1) Add    2) Search    3) Update")
        print("4) Delete 5) Display   6) Exit")
        choice = input("Select option: ").strip()

        if choice == '1':
            add_contact()
        elif choice == '2':
            search_contact()
        elif choice == '3':
            update_contact()
        elif choice == '4':
            delete_contact()
        elif choice == '5':
            display_all()
        elif choice == '6':
            print("Goodbye!")
            break
        else:
            print("Invalid selection. Try again.")


main_menu()
```

Student Hobby Tracker System
Problem Statement:

Create a Student Hobby Tracker System using Python. The system should:

Allow the entry of multiple students and their hobbies.

1.Store each student’s hobbies using a set to avoid duplicates.

2.Provide functionalities using functions to:

Add student hobby data.

Display all student hobbies.

Find students with common hobbies.

List all unique hobbies.



```python
# student_hobby_tracker

hobby_data = {}

def add_student_hobby():
    name = input("Enter student name: ").strip()
    if not name:
        print("Name cannot be empty.")
        return

    hobby = input("Enter a hobby: ").strip()
    if not hobby:
        print("Hobby cannot be empty.")
        return

    if name not in hobby_data:
        hobby_data[name] = set()
    before = len(hobby_data[name])
    hobby_data[name].add(hobby)
    after = len(hobby_data[name])


    if after > before:
        print(f"Hobby '{hobby}' added for {name}.")
    else:
        print(f"'{hobby}' was already listed for {name}.")

def show_all_hobbies():
    if not hobby_data:
        print("No students yet.")
        return
    print("\nAll student hobbies:")
    for student, hobbies in hobby_data.items():
        print(f"{student}: {', '.join(hobbies)}")
    print()

def common_hobbies():
    name1 = input("First student name: ").strip()
    name2 = input("Second student name: ").strip()

    set1 = hobby_data.get(name1)
    set2 = hobby_data.get(name2)

    if not set1 or not set2:
        print("One or both students not found.")
        return

    common = set1.intersection(set2)
    if common:
        print(f"Common hobbies: {', '.join(common)}")
    else:
        print("No common hobbies.")

def list_unique_hobbies():
    all_h = set()
    for hobbies in hobby_data.values():
        all_h.update(hobbies)
    if all_h:
        print("All unique hobbies:")
        for h in all_h:
            print(f"  - {h}")
    else:
        print("No hobbies to list.")

def main_menu():
    while True:
        print("\n----Hobby Tracker Menu----")
        print("1) Add Hobby")
        print("2) Show All")
        print("3) Common Hobbies")
        print("4) List Unique Hobbies")
        print("5) Exit")
        choice = input("Choose an option: ").strip()

        if choice == '1':
            add_student_hobby()
        elif choice == '2':
            show_all_hobbies()
        elif choice == '3':
            common_hobbies()
        elif choice == '4':
            list_unique_hobbies()
        elif choice == '5':
            print("Bye!")
            break
        else:
            print("Invalid choice, please try again.")
main_menu()
```
